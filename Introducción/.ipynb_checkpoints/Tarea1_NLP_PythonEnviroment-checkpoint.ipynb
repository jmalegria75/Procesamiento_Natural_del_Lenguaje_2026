### Usando archivo environment.yml

Cree el siguiente archivo `environment.yml`

```yaml
name: nlp_curso_2026

channels:
  - conda-forge
  - pytorch
  - defaults

dependencies:
  # Python
  - python=3.10.12

  # Jupyter
  - jupyterlab>=4.0
  - ipykernel>=6.29
  - ipywidgets>=8.1

  # NLP clásico
  - nltk>=3.8
  - spacy>=3.7
  - gensim>=4.3

  # Machine Learning
  - scikit-learn>=1.4
  - imbalanced-learn>=0.11

  # Deep Learning
  - pytorch>=2.1
  - torchvision>=0.16
  - cpuonly  # Eliminalo si tienes GPU NVIDIA

  # Data
  - numpy>=1.26
  - pandas>=2.1
  - scipy>=1.11

  # Visualización
  - matplotlib>=3.8
  - seaborn>=0.13
  - plotly>=5.18

  # Utilidades
  - tqdm>=4.66
  - requests>=2.31
  - python-dotenv>=1.0

  # PDF / Archivos
  - pypdf>=3.17

  # Via pip (no disponibles en conda-forge)
  - pip
  - pip:
    - transformers>=4.38
    - datasets>=2.17
    - tokenizers>=0.15
    - evaluate>=0.4
    - accelerate>=0.27
    - sentence-transformers>=2.5
    - huggingface-hub>=0.21
    - fastapi>=0.110
    - uvicorn>=0.27
    - gradio>=4.19
    - python-multipart>=0.0.9
    - pydantic>=2.6
    - wordcloud>=1.9
    - contractions>=0.1
```


#### 1. Instalar el entorno para el curso:


```bash
# 1. Abre Anaconda Prompt (Windows) o Terminal (Mac/Linux)
# 2.  Navega a la carpeta del curso: cd "ruta/al/curso"
# 3. Crea el entorno desde el archivo
conda env create -f environment.yml

# 4. Activa el entorno
conda activate nlp_curso_2026

# 5. Descarga modelos de spaCy
python -m spacy download es_core_news_sm
python -m spacy download en_core_web_sm

# 6. Descarga recursos de NLTK
python -c "import nltk; [nltk.download(r) for r in ['punkt','punkt_tab','stopwords','wordnet','movie_reviews','averaged_perceptron_tagger','maxent_ne_chunker','words']]"

# 7. Registra el kernel en Jupyter
python -m ipykernel install --user --name nlp_curso_2026 --display-name "Python (NLP 2026)"

# 8. Inicia JupyterNotebook
jupyter notebook
```


#### 2. Verifica las librerías instaladas
Crea un nuevo archivo de jupyter para verificar las librerías del entorno (asegurate que el entorno esté activado)
```bash
# Verificación de librerías instaladas
import importlib

librerias = ["nltk", "spacy", "sklearn", "gensim", "numpy", "pandas", "matplotlib"]

print("=" * 45)
print("  VERIFICACIÓN DEL ENTORNO NLP 2026")
print("=" * 45)

for lib in librerias:
    try:
        mod = importlib.import_module(lib)
        version = getattr(mod, "__version__", "N/A")
        print(f"  ✅ {lib:<20} v{version}")
    except ImportError:
        print(f"  ❌ {lib:<20} NO INSTALADO")

print("=" * 45)
```
