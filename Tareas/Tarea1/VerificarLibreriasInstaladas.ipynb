```python
# Verificación de librerías instaladas
import importlib

librerias = ["nltk", "spacy", "sklearn", "gensim", "numpy", "pandas", "matplotlib"]

print("=" * 45)
print("  VERIFICACIÓN DEL ENTORNO NLP 2026")
print("=" * 45)

for lib in librerias:
    try:
        mod = importlib.import_module(lib)
        version = getattr(mod, "__version__", "N/A")
        print(f"  ✅ {lib:<20} v{version}")
    except ImportError:
        print(f"  ❌ {lib:<20} NO INSTALADO")

print("=" * 45)
```
